### Step 6: Explore price by brand.

```python
autos["brand"].value_counts(normalize = True)
```

#### Finding:
There are a lot of brands with volume% < 2%. We will limit our analysis to those brands with vol% > 2%.

#### Action:
- Identify the unique values we want to aggregate by
- Create an empty dictionary to store our aggregate data
- Loop over the unique values, and for each:
    - Subset the dataframe by unique values
    - Calculate the mean of column we're interested in (We will explore `"price"`)
    - Assign the val/mean to the dict as k/v.

```python
brand_counts = autos["brand"].value_counts(normalize = True)
common_brand = brand_counts[brand_counts > 0.02].index
print(common_brand)
```

```python
avg_brand_price = {}

for brand in common_brand:
    
    # Subset the dataframe to be all current brand
    autos_brand = autos[autos["brand"] == brand]
    
    # Calculate avg price of that brand
    avg_price = autos_brand["price"].mean()
    
    # assign the avg_price and its corresponding index into the empty dictionary created
    avg_brand_price[brand] = int(avg_price)
    
avg_brand_price
```

#### Finding:
Mercedes Benz had highest average price of used cars, however Volkswagen is the most popular car brand.

### Step 7: Explore the Mileage

Calculate the average mileage by each brand. Assign both average price and average mileage into one summaried dataframe.

```python
avg_brand_mileage = {}

for brand in common_brand:
    
    # Subset the dataframe to be all current brand
    autos_brand = autos[autos["brand"] == brand]
    
    # Calculate avg price of that brand
    avg_mileage = autos_brand["odometer_km"].mean()
    
    # assign the avg_price and its corresponding index into the empty dictionary created
    avg_brand_mileage[brand] = int(avg_mileage)
    
avg_brand_mileage
```

```python
avg_brand_price_series = pd.Series(avg_brand_price)
avg_brand_mileage_series = pd.Series(avg_brand_mileage)
```

```python
autos_brand_sum = pd.DataFrame(avg_brand_price_series, columns = ['avg_price'])
```

```python
autos_brand_sum['avg_mileage'] = avg_brand_mileage_series
```

```python
autos_brand_sum
```

#### Finding:
The mileage does not differ too much in different bands.
